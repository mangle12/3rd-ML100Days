```python
import pandas as pd
import numpy as np

data_path = 'data/'
df_train = pd.read_csv(data_path + 'titanic_train.csv')
df_test = pd.read_csv(data_path + 'titanic_test.csv')


#train_Y = np.log1p(df_train['SalePrice']) #去掉極端值
train_Y = df_train['Survived']
ids = df_test['PassengerId']
df_train = df_train.drop(['PassengerId', 'Survived'] , axis=1)#刪除Id、SalePrice欄位
df_test = df_test.drop(['PassengerId'] , axis=1)#刪除Id欄位
df = pd.concat([df_train,df_test])#合併兩個dataframe
#print(df.head())

dtype_df = df.dtypes.reset_index()
dtype_df.columns = ["Count", "Column Type"]#放上欄位名稱
dtype_df = dtype_df.groupby("Column Type").aggregate('count').reset_index()#合併相同型態之欄位
#print(dtype_df)

#宣告陣列
int_features = []
float_features = []
object_features = []

for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64':
        float_features.append(feature)
    elif dtype == 'int64':
        int_features.append(feature)
    else:
        object_features.append(feature)

#列出各型態的數量及欄位名稱
#print(f'{len(int_features)} Integer Features : {int_features}\\n')
#print(f'{len(float_features)} Float Features : {float_features}\\n')
#print(f'{len(object_features)} Object Features : {object_features}')
#print(df[float_features].head())#單獨show出float的型態欄位值

#作業一

##int取平均
int_Mean = df[int_features].mean()
print("int平均: \n {}".format(int_Mean))

print()
##int取最大值
int_Max = df[int_features].max()
print("int最大值: \n {}".format(int_Max))

print()
##int取相異值
int_Nunique = df[int_features].nunique()
print("int相異值: \n {}".format(int_Nunique))

print()
##float取相異值
float_Mean = df[float_features].mean()
print("float平均: \n {}".format(float_Mean))

print()
##float取最大值
float_Max = df[float_features].max()
print("float最大值: \n {}".format(float_Max))

print()
##float取相異值
float_Nunique = df[float_features].nunique()
print("float相異值: \n {}".format(float_Nunique))

print()
##object取相異值
object_Mean = df[object_features].mean()
print("object平均: \n {}".format(object_Mean))

print()
##object取最大值
object_Max = df[object_features].max()
print("object最大值: \n {}".format(object_Max))

print()
##object取相異值
object_Nunique = df[object_features].nunique()
print("object相異值: \n {}".format(object_Nunique))

###object資料型態無法取平均值和最大值，所以會出錯


#作業二
#類型:Boolean 、List 、Tuple 、Dictionary 、Sets 
#object類型處理最複雜
```
